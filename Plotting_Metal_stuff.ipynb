### Stellar mass vs metallicity

I want to plot stellar mass on x axis and metallicities on y axis.

```python
halo1 = sim.timesteps[-1][1]
```

```python
step = db.get_timestep('cosmo25/%8192')
stellar_mass, star_metal_prof, halo_num = step.calculate_all('Mstar', 'at(radius(200), star_metal_profile)', 'halo_number()') 
```

```python
print(star_metal_prof)
```

```python
plt.scatter(stellar_mass, star_metal_prof)

plt.xscale('log')
plt.yscale('log')

plt.xlabel('Stellar Mass $[M_{\odot}]$')
plt.ylabel('Star Metal Profile within R = 200kpc of halo')

plt.title('Stellar Mass vs Star Metal Profile')
plt.show()
```

```python
##plotting the cumulative star metal profile of the most massive halo at last timestep

halo1_star_metal = halo1['star_metal_profile']
r_bins = np.arange(len(halo1_star_metal))*0.1+0.05
plt.plot(r_bins, np.cumsum(halo1_star_metal))

plt.xlabel('R [kpc]')
plt.ylabel('star metal profile')

plt.title('Radius vs Star Metal Profile')

plt.show()
```

```python
step = db.get_timestep('cosmo25/%8192')
gas_mass, gas_metal_prof, halo_num = step.calculate_all('Mgas', 'at(radius(200), gas_metal_profile)', 'halo_number()') 
```

```python
ok = np.where(gas_mass > 0)

plt.scatter(gas_mass[ok], gas_metal_prof[ok])

plt.xscale('log')
plt.yscale('log')

plt.xlabel('Gas Mass $[M_{\odot}]$')
plt.ylabel('Gas Metal Profile within R = 200kpc of halo')

plt.title('Gas Mass vs Gas Metal Profile')

plt.show()
```

```python
halo1_gas_metal = halo1['gas_metal_profile']
r_bins = np.arange(len(halo1_gas_metal))*0.1+0.05
plt.plot(r_bins, halo1_gas_metal)

plt.xlabel('R [kpc]')
plt.ylabel('Gas metal profile')

plt.title('Radius vs Gas Metal Profile')

plt.show()
```

```python
step = db.get_timestep('cosmo25/%8192')
star_metal, gas_metal, star_Fe, gas_Fe, halo_num = step.calculate_all('at(radius(200), star_metal_profile)',
                                                     'at(radius(200), gas_metal_profile)',
                                                     'at(radius(200), star_Fe_profile)',
                                                     'at(radius(200), gas_Fe_profile)',
                                                     'halo_number()') 
```

```python
r_bins1 = np.arange(len(star_metal))*0.1+0.05
r_bins2 = np.arange(len(gas_metal))*0.1+0.05

plt.plot(r_bins1, np.cumsum(star_metal), label = 'stellar')
plt.plot(r_bins2, np.cumsum(gas_metal), label = 'gas')

plt.xlabel('R [kpc]')
plt.ylabel('Metal profile')

plt.title('Radius vs Star Metal Profile')

plt.legend()
plt.show()
```

```python
r_bins1 = np.arange(len(star_metal))*0.1+0.05

plt.scatter(r_bins1, star_metal, marker = '*', label = 'stellar',
           c= np.log(star_Fe), cmap = 'viridis', vmin = -15, vmax = -4)

plt.xscale('log')
plt.yscale('log')

plt.xlabel('R [kpc]')
plt.ylabel('Metal profile')

plt.title('Radius vs Star Metal Profile')
plt.colorbar(label='Log Star Fe Profile')

plt.grid()
plt.legend()
plt.show()
```

```python
r_bins2 = np.arange(len(gas_metal))*0.1+0.05

plt.scatter(r_bins2, gas_metal, marker = 'o', label = 'gas',
           c= np.log(gas_Fe), cmap = 'viridis', vmin = -15, vmax = -4)

plt.xscale('log')
plt.yscale('log')

plt.xlabel('R [kpc]')
plt.ylabel('Metal profile')

plt.title('Radius vs Gas Metal Profile')
plt.colorbar(label='Log Gas Fe Profile')

plt.grid()
#plt.legend()
plt.show()
```

```python
stellar = (stellar_mass * stellar_metal)/Mtot
gas = (gas_mass * gas_metal)/Mtot

plt.scatter(stellar_mass,stellar, label = 's')
plt.scatter(gas_mass, gas, label = 'g')

plt.xscale('log')
plt.yscale('log')

plt.legend()
plt.grid()
plt.show()
```
